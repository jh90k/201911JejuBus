```python
import seaborn as sns
```

```python
dfTrain['18~20_ride'].value_counts().plot(kind='bar')
plt.title('number of ppl riding 18-10')
plt.xlabel('passenger')
plt.ylabel('count')
sns.despine
```

```python
#location information
plt.figure(figsize=(15,15))
sns.jointplot(x=dfTrain.latitude.values, y=dfTrain.longitude.values, size=15)
plt.xlabel('Latitude', fontsize=12)
plt.ylabel('Longitude', fontsize=12)
plt.show()
```
